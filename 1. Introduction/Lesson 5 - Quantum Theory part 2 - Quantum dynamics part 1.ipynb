```python
import numpy as np
from scipy.integrate import solve_ivp

def hamiltonian(t, psi):
    return -1j * H @ psi  # Define the Hamiltonian H as a matrix

# Initial state |psi(0)>
psi0 = np.array([1, 0])  # Example: Pure state |0>

# Time points
t_span = (0, 1)  # Time interval from 0 to 1

# Solve the time-dependent Schr√∂dinger equation
sol = solve_ivp(hamiltonian, t_span, psi0)
psi_t = sol.y
```

```python
import numpy as np
from scipy.integrate import solve_ivp

def heisenberg_eqn(t, A):
    A_matrix = A.reshape((2, 2))  # Reshape 1D array back to 2x2 matrix
    commutator = 1j * np.dot(H, A_matrix) - 1j * np.dot(A_matrix, H)
    return commutator.flatten()  # Flatten matrix back to 1D array for solve_ivp

# Define Hamiltonian H and initial operator A(0) as matrices
H = np.array([[1, 0], [0, -1]])  # Example 2x2 Hamiltonian
A0 = np.array([[0, 1], [1, 0]]).flatten()  # Example initial operator, flattened to 1D array

# Constants
hbar = 1.0  # Reduced Planck's constant (modify as needed)

# Time points
t_span = (0, 1)  # Time interval from 0 to 1

# Solve the Heisenberg equation of motion
sol = solve_ivp(heisenberg_eqn, t_span, A0)
A_t = sol.y.reshape((2, 2, len(sol.t)))  # Reshape solution back to array of matrices

# Output
print("Solution at final time:")
print(A_t[:, :, -1])  # Display the final operator at the last time point
```

```python
import numpy as np

# Define constants
hbar = 1.0  # Reduced Planck's constant (modify as needed)
E = 1.0  # Energy eigenvalue

# Define initial state |psi(0)>
psi0 = np.array([1, 0])  # Example: Pure state |0>

# Time points
t = 1.0  # Time at which to evaluate the state

# Compute time-evolved state |psi(t)>
psi_t = np.exp(-1j * E * t / hbar) * psi0

# Print the evolved state
print("Evolved State at t =", t, ":")
print(psi_t)
```
