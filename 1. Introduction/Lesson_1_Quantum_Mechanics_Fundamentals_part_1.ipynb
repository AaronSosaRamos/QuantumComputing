```python
import cirq
import numpy as np

# Define the qubit
qubit = cirq.LineQubit(0)  # Using a single qubit

# Create a quantum circuit
circuit = cirq.Circuit()

# Apply a superposition state (wave-like behavior)
circuit.append(cirq.H(qubit))  # Hadamard gate to create superposition

# Measure the quantum state (particle-like behavior)
circuit.append(cirq.measure(qubit, key='result'))  # Measurement

# Simulate the circuit
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=1000)

# Analyze the measurement results
counts = result.histogram(key='result')
print("Measurement outcome probabilities:", counts)

# Visualize the circuit
print("\nCircuit:")
print(circuit)
```

```python
import cirq

# Define the qubit
qubit = cirq.LineQubit(0)  # Using a single qubit

# Create a quantum circuit
circuit = cirq.Circuit()

# Apply a Hadamard gate to create a superposition state
circuit.append(cirq.H(qubit))

# Measure the quantum state
circuit.append(cirq.measure(qubit, key='result'))

# Display the circuit
print("Quantum Circuit for Superposition with Measurement:")
print(circuit)

# Simulate the circuit
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=1000)

# Analyze the measurement results
counts = result.histogram(key='result')

# Print the measurement outcome probabilities
print("\nMeasurement Outcome Probabilities:")
for outcome in counts:
    probability = counts[outcome] / 1000.0  # Normalize to probability
    print(f"Outcome {outcome}: Probability = {probability:.3f}")
```

#Multi-Qubit Superposition Circuit


```python
import cirq

# Define two qubits
qubit1 = cirq.LineQubit(0)
qubit2 = cirq.LineQubit(1)

# Create a quantum circuit
circuit = cirq.Circuit()

# Apply Hadamard gates to create superposition on each qubit
circuit.append(cirq.H(qubit1))
circuit.append(cirq.H(qubit2))

# Apply a controlled-X (CNOT) gate
circuit.append(cirq.CNOT(qubit1, qubit2))

# Measure the qubits
circuit.append(cirq.measure(qubit1, key='result1'))
circuit.append(cirq.measure(qubit2, key='result2'))

# Display the circuit
print("Multi-Qubit Superposition Circuit:")
print(circuit)

# Simulate the circuit
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=1000)

# Extract measurement results
measurements = result.measurements
results1 = measurements['result1']
results2 = measurements['result2']

# Count measurement outcomes
counts = {}
for r1, r2 in zip(results1, results2):
    outcome = (r1[0], r2[0])  # Combine outcomes of qubit1 and qubit2
    if outcome in counts:
        counts[outcome] += 1
    else:
        counts[outcome] = 1

# Print the measurement outcome probabilities
print("\nMeasurement Outcome Probabilities:")
total_samples = sum(counts.values())
for outcome, count in counts.items():
    probability = count / total_samples
    print(f"Outcome {outcome}: Probability = {probability:.3f}")
```

```python
import cirq

# Define two qubits
qubit1 = cirq.LineQubit(0)
qubit2 = cirq.LineQubit(1)

# Create a quantum circuit
circuit = cirq.Circuit()

# Apply a Hadamard gate to qubit1 to create a superposition state
circuit.append(cirq.H(qubit1))

# Apply a CNOT gate with qubit1 as control and qubit2 as target
circuit.append(cirq.CNOT(qubit1, qubit2))

# Display the circuit
print("Quantum Circuit for Entanglement:")
print(circuit)

# Simulate the circuit
simulator = cirq.Simulator()
result = simulator.simulate(circuit)

# Extract the final state vector
state_vector = result.final_state_vector

# Print the final state vector
print("\nFinal State Vector:")
print(state_vector)
```
